```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
df = pd.read_csv('Titanic.csv')
```

```python
df.head
```

```python
survival_rate = df.groupby('Sex')['Survived'].mean()*100
plt.bar(survival_rate.index,survival_rate.values,color='blue')
plt.title('Survival rate per gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
```

```python
survival_rate_age = df.groupby('Age')['Survived'].mean()*100
plt.hist(survival_rate_age,bins=20)
plt.title('Survival rate age')
plt.xlabel('age')
plt.ylabel('Survival Rate')
```
